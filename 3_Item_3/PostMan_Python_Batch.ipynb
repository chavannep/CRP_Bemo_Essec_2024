# Batch computation 

## Over price formula with unique inputs

```python
df_results = pd.DataFrame.from_dict({'startDateTime_var':[],\
          'sessionVolumme_var':[],\
          'sessionType_var':[],\
          'partnerCdrId_var':[],\
          'currency_var':[],\
          'elements_var':[],\
          'country_var':[],\
          'body_dict':[],\
          'computation_successful':[],\
          'time_price':[],\
          'energy_price':[],\
          'flat_price':[],\
          'parking_price':[],\
          'total_price':[]})

# Looping through price formula with unique above inputs
for i in range(df_price_formula.shape[0]):
    
    formula_dict = ast.literal_eval(df_price_formula['Formula'].loc[i])
    partnerCdrId_var = df_price_formula['Id'].loc[i]
    currency_var = formula_dict['currency']
    elements_var = formula_dict['elements']
    country_var = df_price_formula['Country'].loc[i]
    
    body_dict = {"session":{"chargingPeriods": [{"startDateTime": startDateTime_var,\
                                                 "dimensions": [{"type": sessionType_var,\
                                                                 "volume": sessionVolumme_var}]}],\
                            "partnerCdrId": partnerCdrId_var},\
                 "tariff": {"currency": currency_var,\
                            "elements": elements_var,\
                            "countryCode": country_var}\
                }
    
    url = "https://dev-pricer.be-mo.io/api/price/ocpi2-2-1/eichrecht-simple"
    response = requests.request("POST",\
                                url,\
                                headers={'Content-Type': 'application/json'},\
                                data=json.dumps(body_dict),\
                                verify=False)
    response_dict = response.json()
    
    # Computation successful : exporting results
    if('errors' not in list(response_dict.keys())):
        if(response_dict['time'] == None):
            time_price = 0
        else:
            time_price = response_dict['time']
            
        if(response_dict['energy'] == None):
            energy_price = 0
        else:
            energy_price = response_dict['energy']    
    
        if(response_dict['flat'] == None):
            flat_price = 0
        else:
            flat_price = response_dict['flat']   
    
        if(response_dict['parking_time'] == None):
            parking_price = 0
        else:
            parking_price = response_dict['parking_time']
    
        total_price = time_price + energy_price + flat_price + parking_price
        computation_successful = 1
        
    # At least one error occured : writing NaNs
    else:
        time_price = np.nan
        energy_price = np.nan
        flat_price = np.nan
        parking_price = np.nan
        total_price = np.nan
        computation_successful = 0
    
    new_row = pd.DataFrame.from_dict({'startDateTime_var':[startDateTime_var],\
          'sessionVolumme_var':[sessionVolumme_var],\
          'sessionType_var':[sessionType_var],\
          'partnerCdrId_var':[partnerCdrId_var],\
          'currency_var':[currency_var],\
          'elements_var':[elements_var],\
          'country_var':[country_var],\
          'body_dict':[body_dict],\
          'computation_successful':[computation_successful],\
          'time_price':[time_price],\
          'energy_price':[energy_price],\
          'flat_price':[flat_price],\
          'parking_price':[parking_price],\
          'total_price':[total_price]})
    df_results = pd.concat([df_results, new_row], axis=0) #join='outer', ignore_index=False)

# Saving all results in a CSV file
df_results.to_csv('./Computation_results_batchPriceFormula.csv', index=False, index_label=True, header=True, encoding='utf-8', quotechar='"', decimal='.', sep=';') # encoding='iso-8859-1'
```
